```python
import pandas as pd
import re
```

```python
def predicting_txt_to_dataframe(path):    
    with open(path, 'r') as f:
        txt = str(f.read())
        txt_list = txt.split('\n')
    row_list = list()
    tmp_list = list()
    for row in txt_list:
        if row == '--------------------':
            tmp_list[0] = tmp_list[0].replace('article_id:', '')
            tmp_list[0] = int(tmp_list[0])
            row_list.append(tmp_list)
            tmp_list = list()
            continue
        if len(row) > 0:
            tmp_list.append(row)
    df = pd.DataFrame(row_list, columns=['article_id','text'])
    return df
```

```python
df = predicting_txt_to_dataframe('/home/Danny/AI-CUP-2020/datasets/stage5/test.txt')
df
```

```python
def get_identity_ID(article_id, text):
    output_str = str()
    entity_type = 'ID'
    source = '1-re'
    for match in re.finditer('[a-zA-Z]{1}[0-9]{9}', text):
        output_str += '{}\t{}\t{}\t{}\t{}\t{}\n'.format(article_id, match.start(), match.end(), match.group(), entity_type, source)
    return output_str
```

```python
def get_phone_number(article_id, text):
    output_str = str()
    entity_type = 'contact'
    source = '1-re'
    for match in re.finditer('09[0-9]{8}', text):
        output_str += '{}\t{}\t{}\t{}\t{}\t{}\n'.format(article_id, match.start(), match.end(), match.group(), entity_type, source)
    return output_str
```

```python
def get_ip_address(article_id, text):
    output_str = str()
    entity_type = 'contact'
    source = '1-re'
    for match in re.finditer('www[.[a-zA-Z]*]*', text):
        output_str += '{}\t{}\t{}\t{}\t{}\t{}\n'.format(article_id, match.start(), match.end(), match.group(), entity_type, source)
    return output_str
```

```python
def get_date(article_id, text):
    output_str = str()
    entity_type = 'time'
    source = '1-re'
    for match in re.finditer('[0-9]{1,2}/[0-9]{1,2}', text):
        output_str += '{}\t{}\t{}\t{}\t{}\t{}\n'.format(article_id, match.start(), match.end(), match.group(), entity_type, source)
    return output_str
```

```python
output_str = '{}\t{}\t{}\t{}\t{}\t{}\n'.format('article_id', 'start_position', 'end_position', 'entity_text', 'entity_type', 'source')
for article_id, text in zip(df['article_id'], df['text']):
    output_str += get_identity_ID(article_id, text)
    output_str += get_phone_number(article_id, text)
    output_str += get_ip_address(article_id, text)
    output_str += get_date(article_id, text)
print(output_str)
```

```python
with open('regular_expression.tsv', 'w', encoding='utf-8') as f:
    f.write(output_str)
```
