## Generate clinical medicine

```python
clinical_data = pd.read_csv('./clinical_event.csv')
clinical_data
```

```python
clinical_list = []

with open('疾病列表.txt','r') as f:
    lines = f.readlines()
    f.close()
    
for line in lines:
    print(line)
    results = line.split('	')
    print(results)
    clinical_list.append(results[0].rstrip())
```

```python
clinical_list
```

```python
clinical_obj = []
for disease in clinical_list:
    if 'disease' in disease or 'flu' in disease or 'syndrome' in disease:
        clinical_obj.append(disease.split(' ')[0])
```

```python
clinical_obj
```

```python
clinical_enent_dictionary = pd.DataFrame(clinical_list+clinical_obj,columns=['entity_text'])
clinical_enent_dictionary['entity_type'] = 'clinical_event'
clinical_enent_dictionary
```

```python
clinical_enent_dictionary.to_csv('./dictionaries/clinical_event_dictionary.csv',index=False)
```

## Education 

```python
education_list = []

with open('台灣大學列表.txt','r') as f:
    lines = f.readlines()
    f.close()
    
for line in lines:
    results = line.split('	')
    if len(results) > 3:
        education_list.append(results[1].rstrip())
education_list
```

```python
education_list_obj = []

for school in education_list:
    if '立' in school:
        education_list_obj.append(school.split('立')[1])
education_list_obj
```

```python
education_list_obj.remove('大學')
education_list_obj
```

```python
education_list_obj_v2 = []
for school in education_list:
    if '臺' in school:
        education_list_obj_v2.append(school.replace('臺','台'))
        
education_list_obj_v2
```

```python
education_list_obj_v3 = []
for school in education_list_obj:
    if '臺' in school:
        education_list_obj_v3.append(school.replace('臺','台'))
        
education_list_obj_v3
```

```python
education_list_obj_v4 = []

for school in education_list_obj:
    if '臺' not in school:
        education_list_obj_v4.append(school.split('大學')[0])
        
education_list_obj_v4
```

```python
location = ['宜蘭',
 '體育','嘉義',
 '高雄',
 '金門',
 '屏東','空中']
for loc in location:
    education_list_obj_v4.remove(loc)
            
education_list_obj_v4
```

```python
education_list_obj_v5 = ['台大','臺大','成大','政大','交大','清大','劍橋大學','哈佛','加州柏克萊','港大']

subject_list = []

with open('./系所.txt','r') as f:
    lines = f.readlines()
    f.close()
for line in lines:
    if line[-2] == '系':
        subject_list.append(line.strip('\n'))
        
subject_list
```

```python
subject_list_v2 = []

for subject in subject_list:
    if len(subject) > 3 and '文學' not in subject:
        temp = subject[:-2] + '系'
        subject_list_v2.append(temp)
        
for subject in ['中文系','外文系','日文系','電機系','資工系','化工系','土木系','機械系','資管系','材料系','地理系']:
    subject_list_v2.append(subject)
subject_list_v2
```

```python
subject_list_v3 = []

for subject in subject_list_v2:
    subject_list_v3.append(subject.strip('系'))

for subject in subject_list:
    if len(subject.strip('學系')) >= 2:
        subject_list_v3.append(subject.strip('學系'))
subject_list_v3
```

```python
from itertools import product

school_subject = []

for item in list(product(education_list,subject_list_v3)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj,subject_list_v3)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj_v2,subject_list_v3)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj_v3,subject_list_v3)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj_v4,subject_list_v3)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj_v5,subject_list_v3)):
    school_subject.append(item[0]+item[1])
    
for item in list(product(education_list,subject_list_v2)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj,subject_list_v2)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj_v2,subject_list_v2)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj_v3,subject_list_v2)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj_v4,subject_list_v2)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj_v5,subject_list_v2)):
    school_subject.append(item[0]+item[1])
    
    
for item in list(product(education_list,subject_list)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj,subject_list)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj_v2,subject_list)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj_v3,subject_list)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj_v4,subject_list)):
    school_subject.append(item[0]+item[1])
for item in list(product(education_list_obj_v5,subject_list)):
    school_subject.append(item[0]+item[1])
len(school_subject)
```

```python
total_education = education_list+education_list_obj+education_list_obj_v2+education_list_obj_v3+education_list_obj_v4 + subject_list + subject_list_v2 + subject_list_v3 + school_subject
```

## Profession

```python
profession = pd.read_csv('./profession.csv')
profession
```

```python
profession_list = profession['entity_text'].unique().tolist()
```

```python
profession_list.append('工程士')
profession_list.append('飲料鋪')
profession_list.append('正教授')
profession_list.append('副教授')
profession_list.append('助理教授')
profession_list.append('公務員')
profession_list.append('保險員')
```

```python
with open('職業.txt','r') as f:
    lines = f.readlines()
    f.close()
for line in lines:
    results = line.split(' ')
    if len(results) >= 3:
        temp = results[-2]
        if temp not in profession_list:
            if '（' in temp:
                temp = temp.split('（')[0]
            elif '(' in temp:
                temp = temp.split('(')[0]
            if temp not in ['1','2','3','4','5','6','7','8','9','0'] and temp != '' and temp != '1900002' and ')' not in temp and '）' not in temp and temp != '醫師' and temp != '醫生' and temp != '個管師': 
                for t in temp.split('、'):
                    profession_list.append(t)
profession_list
```

```python
profession_output = pd.DataFrame(profession_list, columns=['entity_text'])
```

```python
profession_output['entity_type'] = 'profession'
```

```python
profession_output
```

```python
profession_output.to_csv('./dictionaries/profession_dictionary.csv',index=False)
```
