```python
import pandas as pd
```

```python
with open('stage1/SampleData_deid.txt', 'r') as f:
    txt = str(f.read())
```

```python
txt_list = txt.split('\n')
```

```python
X_Y_list = list()
tmp = list()
for line in txt_list:
    if line == '--------------------':
        X_Y_list.append(tmp)
        tmp = list()
        continue
    if line == '':
        continue
    tmp.append(line)
```

```python
data_list = list()
for X_Y in X_Y_list:
    text = X_Y[0]
    labels = X_Y[2:]
    for label in labels:
        row_list = label.split('\t')
        row_list.insert(0, text)        
        data_list.append(row_list)
```

```python
df = pd.DataFrame(data_list, columns = ['text', 'article_id', 'start_position', 'end_position', 'entity_text', 'entity_type'])
```
