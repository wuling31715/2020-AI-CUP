```python
import pandas as pd
```

```python
ensemble_df = pd.DataFrame(columns=['article_id', 'start_position', 'end_position', 'entity_text', 'entity_type', 'source'])
ensemble_df
```

```python
bert_df = pd.read_csv('/home/Danny/AI-CUP-2020/danny/bert/output/bert-chinese_epoch_512_batch_1024.tsv', sep='\t')
bert_df['source'] = '3-bert'
bert_df
```

```python
bert_df = bert_df.drop(bert_df[bert_df['entity_type'] == 'ID'].index)
bert_df = bert_df.drop(bert_df[bert_df['entity_type'] == 'location'].index)
bert_df = bert_df.drop(bert_df[bert_df['entity_type'] == 'contact'].index)
bert_df
```

```python
re_df = pd.read_csv('/home/Danny/AI-CUP-2020/regular_expression/regular_expression.tsv', sep='\t')
re_df
```

```python
dict_df = pd.read_csv('/home/Danny/AI-CUP-2020/dictionary/dictionary.tsv', sep='\t')
dict_df
```

```python
ensemble_df = ensemble_df.append(bert_df, ignore_index=True)
ensemble_df = ensemble_df.append(re_df, ignore_index=True)
ensemble_df = ensemble_df.append(dict_df, ignore_index=True)
ensemble_df
```

```python
ensemble_df = ensemble_df.sort_values(by=['article_id', 'start_position', 'source'])
ensemble_df
```

```python
ensemble_df.to_csv('ensemble/ensemble_all.tsv', sep='\t', index=False)
entity_type_list = pd.unique(ensemble_df['entity_type']).tolist()
for entity_type in entity_type_list:
    bert_df[bert_df['entity_type'] == entity_type].to_csv('bert/{}.tsv'.format(entity_type), sep='\t', index=False)
```
