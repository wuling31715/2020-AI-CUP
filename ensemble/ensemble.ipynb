```python
ensemble_df = pd.DataFrame(columns=['article_id', 'start_position', 'end_position', 'entity_text', 'entity_type', 'source'])
ensemble_df
```

```python
bert_df = pd.read_csv('/home/Danny/AI-CUP-2020/danny/bert/output/bert-chinese_epoch_512_batch_1024.tsv', sep='\t')
bert_df['source'] = 'bert'
bert_df
```

```python
regular_expression_df = pd.read_csv('/home/Danny/AI-CUP-2020/regular_expression/regular_expression.tsv', sep='\t')
regular_expression_df
```

```python
dictionary_df = pd.read_csv('/home/Danny/AI-CUP-2020/dictionary/dictionary.tsv', sep='\t')
dictionary_df
```
