```python
import pandas as pd
```

```python
def predicting_txt_to_dataframe(path):    
    with open(path, 'r') as f:
        txt = str(f.read())
        txt_list = txt.split('\n')
    row_list = list()
    tmp_list = list()
    for row in txt_list:
        if row == '--------------------':
            tmp_list[0] = tmp_list[0].replace('article_id:', '')
            tmp_list[0] = int(tmp_list[0])
            row_list.append(tmp_list)
            tmp_list = list()
            continue
        if len(row) > 0:
            tmp_list.append(row)
    df = pd.DataFrame(row_list, columns=['article_id','text'])
    return df
```

```python
text_df = predicting_txt_to_dataframe('/home/Danny/ai-cup-2020/datasets/stage5/test.txt')
text_df
```

```python
label_df = pd.read_csv('/home/Danny/ai-cup-2020/ensemble/ensemble/ensemble_all.tsv', sep='\t')
label_df
```

```python
label_list_list = list()
for article_id, start_position, end_position, entity_type in zip(label_df['article_id'], 
                                                                 label_df['start_position'],
                                                                 label_df['end_position'],
                                                                 label_df['entity_type'],
                                                                 ):
    
    label_list = [article_id, start_position, end_position, entity_type]
    label_list_list.append(label_list)
```

```python
label_list_list_list = [[] for i in range(159)]
for label_list in label_list_list:
    article_id = label_list[0]
    label_list = label_list[1:]    
    label_list_list_list[article_id].append(label_list)
```

```python
text_df['labels'] = label_list_list_list
text_df
```

```python
text_df = text_df.drop(columns=['article_id'])
text_df.to_csv('ai_cup.csv', index=False)
```
