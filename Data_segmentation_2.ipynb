```python
# 不用儲存article_id, 轉成list
def training_txt_to_list(path):
    with open(path, 'r') as f:
        txt = str(f.read())
    txt_list = txt.split('\n')
    text_label_list = list()
    tmp = list()
    for line in txt_list:
        if line == '--------------------':
            text_label_list.append(tmp)
            tmp = list()
            continue
        if line == '':
            continue
        tmp.append(line)
    x = list()
    y = list()
    for text_label in text_label_list:
        text = text_label[0]
        label = text_label[2:]
        label_list = ['O' for i in range(len(text))]
        for i in label:
            entity = i.split('\t')
            if int(entity[1]) > int(entity[2]):
                continue
            b = int(entity[1])
            label_list[b] = 'B-{}'.format(entity[-1])
            for j in range(int(entity[1])+1, int(entity[2])):
                label_list[j] = 'I-{}'.format(entity[-1])
            
        for i, j in enumerate(text):
            if j == '，' or j == '。' or j == '？':
                label_list[i] = j
        str1_1="民"
        str1_2="眾"
        str2_1="醫"
        str2_2="師"
        str3_1="家"
        str3_2="屬"
        str4_1="個"
        str4_2="管"
        str4_3="師"
        str5_1="護"
        str5_2="理"
        str5_3="師"
        positions=[]
        for j in range(len(text)):
            if j!=(len(text)-1):
                if text[j]==str1_1 and text[j+1]==str1_2:
                    positions.append(j)
                elif text[j]==str2_1 and text[j+1]==str2_2:
                    positions.append(j)
                elif text[j]==str3_1 and text[j+1]==str3_2:
                    positions.append(j)
                elif text[j]==str4_1 and text[j+1]==str4_2 and text[j+1]==str4_3:
                    positions.append(j)
                elif text[j]==str5_1 and text[j+1]==str5_2 and text[j+1]==str5_3:
                    positions.append(j)

        for index in range(len(positions)):
            if index!=(len(positions)-1):
                start_index=positions[index]
                end_index=positions[index+1]
                sentence=text[start_index:end_index]
                y_sentence=label_list[start_index:end_index]
            else:
                start_index=positions[index]
                sentence=text[start_index:]
                y_sentence=label_list[start_index:]
            x.append(sentence)
            y.append(y_sentence)
        print(x,y)
    return [x, y]
```
